```python
from astropy.io import fits
from astropy.table import Table
from astroquery.gaia import Gaia
import pandas as pd
import re
```

```python
HMXBtable = Table.read('HMXB_cat.fits')
LMXBtable = Table.read('LMXB_cat.fits')
```

```python
HMXBtable = HMXBtable.to_pandas()
```

```python
HMXBtable.sort_values(by='PosErr')
```

```python
LMXBtable = LMXBtable.to_pandas()
```

```python
HMXBtable.sort_values(by='PosErr')
```

```python

# Create an empty list to hold all result rows
combined_results = []

def process_table(df, label):
    for idx, row in df.iterrows():
        pos_err = row['PosErr']

        # Skip rows with excessive uncertainty
        if pos_err > 0.05:  # or 179 for stricter filtering
            continue

        name = str(row['Name'])
        ra = row['RAdeg']
        dec = row['DEdeg']

        safe_name = re.sub(r'\W+', '_', name)
        source_id = f"{label.lower()}{idx}_{safe_name}"

        query = f"""
        SELECT *
        FROM gaiadr3.gaia_source
        WHERE 1 = CONTAINS(
            POINT('ICRS', ra, dec),
            CIRCLE('ICRS', {ra}, {dec}, {pos_err})
        )
        AND phot_g_mean_mag < 20
        AND parallax IS NOT NULL
        """

        job = Gaia.launch_job_async(query)
        results = job.get_results()
        results_df = results.to_pandas()

        # Add source info to each row before merging
        results_df['SourceType'] = label
        results_df['SourceID'] = source_id
        results_df['Catalog_RA'] = ra
        results_df['Catalog_Dec'] = dec
        results_df['Catalog_PosErr'] = pos_err

        combined_results.append(results_df)

# 🔄 Run for both tables
process_table(HMXBtable, "HMXB")
process_table(LMXBtable, "LMXB")

# 🧾 Concatenate all results into one big DataFrame
final_df = pd.concat(combined_results, ignore_index=True)

# 💾 Save to a master CSV file
final_df.to_csv("combined_gaia_results.csv", index=False)
```
